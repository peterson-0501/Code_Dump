**a. Marginal PDFs of X and Y:**


We start with the given joint probability density function (PDF):

$$
f_{X,Y}(x,y) =
\begin{cases} 
\frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}}, & \text{if } x > 0, \ 0 < y < 1, \\
0, & \text{otherwise}.
\end{cases}
$$


The marginal PDF of $ X$ is obtained by integrating out $ y $:

$$
f_X(x) = \int_0^1 f_{X,Y}(x,y) \, dy
$$

Substituting the given joint PDF:

$$
f_X(x) = \int_0^1 \frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}} \, dy
$$


Let:

$$
t = \frac{x}{y^{1/\alpha}}
$$

Differentiating both sides:

$$
\frac{x}{y^{1/\alpha + 1}} dy = -\alpha dt
$$

Also, the limits change as follows:
- When $y \to 0 $, then $t \to \infty $.
- When $y \to 1 $, then $ t \to x $.


Rewriting the integral in terms of $ t $:

$$
f_X(x) = \frac{-\alpha x^{\alpha - 1}}{\Gamma(\alpha + 1)} \int_{\infty}^{x} e^{-t} dt
$$

Simplifying:

$$
f_X(x) = \frac{x^{\alpha -1}}{\Gamma(\alpha)} \int_x^\infty  e^{-t} dt
$$

we obtain:

$$
f_X(x) = \frac{x^{\alpha -1}}{\Gamma(\alpha)} e^{-x}
$$
Hence we got,


$$
f_X(x) = 
\begin{cases}
\frac{x^{\alpha -1} e^{-x}}{\Gamma(\alpha)}, & 0 < x < \infty, \\
0, & \text{otherwise}
\end{cases}
$$


The above expression is the probability density function of a **Gamma distribution** with shape parameter $\alpha$ and scale parameter $ 1 $:

$$
X \sim \text{Gamma}(\alpha, 1).
$$



```python
# Marginal PDFs of f_X(x)

import sympy as sp

# Define symbols
x, y, alpha = sp.symbols('x y alpha', positive=True, real=True)

# Given joint PDF
joint_pdf = (1 / sp.gamma(alpha + 1)) * (x**alpha / y) * sp.exp(-x / y**(1/alpha)) * (1 / y**(1/alpha))

# Compute marginal PDF of X
marginal_X = sp.integrate(joint_pdf, (y, 0, 1))

# Simplify result
marginal_X_simplified = sp.simplify(marginal_X)

# Display the result
marginal_X_simplified
```

We start with the given joint probability density function (PDF):

$$
f_{X,Y}(x,y) =
\begin{cases} 
\frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}}, & \text{if } x > 0, \ 0 < y < 1, \\
0, & \text{otherwise}.
\end{cases}
$$


The marginal PDF of $ Y$ is obtained by integrating out $ x $:

$$
f_Y(y) = \int_0^\infty f_{X,Y}(x,y) \, dx
$$

Substituting the given joint PDF:

$$
f_Y(y) = \int_0^\infty \frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}} \, dx
$$

Let:

$$
t = \frac{x}{y^{1/\alpha}}
$$

Differentiating both sides:

$$
\frac{dx}{y^{1/\alpha}} = dt
$$


Also, the limits change as follows:
- When $x \to 0 $, then $t \to 0 $.
- When $x \to \infty $, then $ t \to \infty $.



We reached the integral:

$$
f_Y(y) = \frac{1}{\Gamma(\alpha + 1) y} \int_0^\infty (t y^{1/\alpha})^\alpha e^{-t} dt
$$

Rewriting:

$$
f_Y(y) = \frac{y^{\alpha / \alpha}}{\Gamma(\alpha + 1) y} \int_0^\infty t^\alpha e^{-t} dt
$$

Since $ y^{\alpha / \alpha} = y $, we simplify:

$$
f_Y(y) = \frac{y}{\Gamma(\alpha + 1) y} \int_0^\infty t^\alpha e^{-t} dt
$$

Since the integral is $ \Gamma(\alpha + 1)$, we get:

$$
f_Y(y) = \frac{y}{\Gamma(\alpha + 1) y} \Gamma(\alpha + 1)
$$

$$
f_Y(y) = 1, \quad 0 < y < 1
$$

Thus, the marginal PDF of $ Y $ is:

$$
f_Y(y) =
\begin{cases}
1, & 0 < y < 1, \\
0, & \text{otherwise}.
\end{cases}
$$
This confirms that $ Y \sim \text{Uniform}(0,1) $.



```python
# Marginal PDFs of f_Y(y)

import sympy as sp

# Define symbols
x, y, alpha = sp.symbols('x y alpha', positive=True, real=True)

# Given joint PDF
joint_pdf = (1 / sp.gamma(alpha + 1)) * (x**alpha / y) * sp.exp(-x / y**(1/alpha)) * (1 / y**(1/alpha))

# Compute marginal PDF of Y
marginal_Y = sp.integrate(joint_pdf, (x, 0, sp.oo))

# Simplify result
marginal_Y_simplified = sp.simplify(marginal_Y)

marginal_Y_simplified
```

**b. Conditional PDFs $f_{X|Y}(x|y)$ and $f_{Y|X}(y|x)$:**

The conditional probability density function (PDF) of \( X \) given \( Y = y \) is defined as:  

$$
f_{X|Y}(x | y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}
$$  

Since we have already derived $f_Y(y)$, which is:  

$$
f_Y(y) =
\begin{cases}
1, & 0 < y < 1, \\
0, & \text{otherwise}.
\end{cases}
$$  

Substituting the given joint PDF:

$$
f_{X|Y}(x | y) = \frac{\frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}}}{1}
$$  

Simplifying:

$$
f_{X|Y}(x | y) = \frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}}, \quad x > 0, \ 0 < y < 1.
$$  



The conditional probability density function (PDF) of $ Y $ given $ X = x $ is defined as:  

$$
f_{Y|X}(y | x) = \frac{f_{X,Y}(x,y)}{f_X(x)}
$$  

Since we have already derived \( f_X(x) \), which is:  

$$
f_X(x) = \frac{x^{\alpha -1} e^{-x}}{\Gamma(\alpha)}
$$  

Substituting the given joint PDF:

$$
f_{Y|X}(y | x) = \frac{\frac{x^\alpha}{\Gamma(\alpha + 1) y^{1 + 1/\alpha}} e^{-x / y^{1/\alpha}}}{\frac{x^{\alpha -1} e^{-x}}{\Gamma(\alpha)}}
$$  

Simplifying:

$$
f_{Y|X}(y | x) = \frac{x e^{-x / y^{1/\alpha}}}{\alpha \ y^{1 + 1/\alpha} e^{-x}}
$$  

Rearranging further:

$$
f_{Y|X}(y | x) = \frac{x}{\alpha y^{1 + 1/\alpha}} e^{x (1 - 1 / y^{1/\alpha})}, \quad 0 < y < 1.
$$  



```python
# conditional pdf x given y 

import sympy as sp

# Define symbols
x, y, alpha = sp.symbols('x y alpha', positive=True, real=True)

# Given joint PDF
joint_pdf = (1 / sp.gamma(alpha + 1)) * (x**alpha / y) * sp.exp(-x / y**(1/alpha)) * (1 / y**(1/alpha))

# Compute marginal PDFs
marginal_X = sp.integrate(joint_pdf, (y, 0, 1))
marginal_Y = sp.integrate(joint_pdf, (x, 0, sp.oo))

# Compute conditional PDFs
conditional_Y_given_X = joint_pdf / marginal_X  # f(y | x) = f(x, y) / f_X(x)
conditional_X_given_Y = joint_pdf / marginal_Y  # f(x | y) = f(x, y) / f_Y(y)

# Simplify results
conditional_Y_given_X_simplified = sp.simplify(conditional_Y_given_X)
conditional_X_given_Y_simplified = sp.simplify(conditional_X_given_Y)

# Display results
print("Conditional PDF of Y given X:")
conditional_Y_given_X_simplified


```

**c. For different choices of $\alpha$, plot the surface of the joint PDF $f(x,y)$ on its support:**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import scipy.special

# Define the joint PDF function
def joint_pdf(x, y, alpha=2):
    if x > 0 and 0 < y < 1:
        return (1 / scipy.special.gamma(alpha + 1)) * (x**alpha / y) * np.exp(-x / y**(1/alpha)) * (1 / y**(1/alpha))
    else:
        return 0

# Create a grid of x and y values
x_vals = np.linspace(0.01, 5, 100)  # Avoid zero to prevent division issues
y_vals = np.linspace(0.01, 0.99, 100)  # Avoid zero and one to stay in support
X, Y = np.meshgrid(x_vals, y_vals)

# Choose different values of alpha
alpha_values = [1,2,3,5,10]

fig, axes = plt.subplots(1, len(alpha_values), figsize=(18, 6), subplot_kw={'projection': '3d'})

for ax, alpha in zip(axes, alpha_values):
    Z = np.array([[joint_pdf(x, y, alpha) for x, y in zip(row_x, row_y)] for row_x, row_y in zip(X, Y)])

    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_title(f'Joint PDF for α = {alpha}')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('f(X, Y)')

plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

# since we have the closed form conditional pdf we will use it or else we can use the joint pdf also 
# we will do both

def f_x_given_y(x,y,alpha=2):
    if x > 0 and 0 < y < 1:
        return f_X_given_Y(x,y,alpha)
    else:
        return 0

def f_y_given_x(x,y,alpha=2):
    if x > 0 and 0 < y < 1:
        return f_Y_given_X(x,y,alpha)
    else:
        return 0

    
# Parameters
M = 5000  # Number of samples
burn_in = 500  # Number of burn-in samples to discard
# Initia1ize x and y
x = [5]  # Initial value for x
y = [0.01]  # Initial value for y

# we will take alpha as 2

# Metropolis-Hastings sampling
for i in range(1, M):
    # Sample x given y (f(x|y))
    z = np.random.exponential()  # Propose new x  # by defualt scale is one and size is one
    u = np.random.uniform(0, 1)  # Uniform random number for acceptance

    # Compute acceptance ratio for x
    nume =  f_x_given_y(z,y[i-1]) * st.expon.pdf(x = x[i-1], scale=1)
    demo = f_x_given_y(x[i-1],y[i-1]) * st.expon.pdf(x = z, scale=1)
    alpha = min(1, nume / demo)  # Acceptance probability

    # Accept or reject the proposal for x
    if u <= alpha:
        x.append(z)
    else:
        x.append(x[i-1])

    # Sample y given x (f(y|x))
    z1 = np.random.uniform(0,1) # Propose new y (extract scalar)
    u1 = np.random.uniform(0, 1)  # Uniform random number for acceptance

    # Compute acceptance ratio for y
    nume1 = f_y_given_x(x[i],z1) * st.uniform.pdf(x = y[i-1],loc = 0,scale = 1) 
    demo1 = f_y_given_x(x[i], y[i-1]) * st.uniform.pdf(x = z1,loc = 0,scale = 1)
    alpha1 = min(1, nume1 / demo1)  # Acceptance probability

    # Accept or reject the proposal for y
    if u1 <= alpha1:
        y.append(z1)
    else:
        y.append(y[i-1])  # Append the current value if rejected

# Discard burn-in samples
x = x[burn_in:]
y = y[burn_in:]

# Create a figure with subplots
fig = plt.figure(figsize=(18, 8))

# Subplot 1: Trace plots of x and y
ax1 = fig.add_subplot(121)
M_list = list(range(len(x)))
ax1.plot(M_list, x, color="blue", label="$f(x|y)$")
ax1.plot(M_list, y, color="red", label="$f(y|x)$")
ax1.set_xlabel("Iteration")
ax1.set_ylabel("Value")
ax1.set_title("Trace Plots of X and Y (MH Algorithm)")
ax1.legend()
ax1.grid()

# Subplot 2: 3D histogram of x and y
ax2 = fig.add_subplot(122, projection='3d')

# Create a 2D histogram
hist, xedges, yedges = np.histogram2d(x, y, bins=30, density=True)

# Create grid for 3D plot
xpos, ypos = np.meshgrid(xedges[:-1] +0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.flatten()
ypos = ypos.flatten()
zpos = np.zeros_like(xpos)

# Define the size of the bars
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.flatten()

# Plot the 3D histogram
ax2.bar3d(xpos, ypos, zpos, dx, dy, dz, alpha=0.7, color='lightblue')

# Add labels and title
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Density')
ax2.set_title('3D Histogram of X and Y (MH Algorithm)')

plt.tight_layout()
plt.show()
```
