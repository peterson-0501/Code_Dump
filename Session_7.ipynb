### a)

As the first experiment, we consider the confidence interval that we have commonly seen in most of 
the textbook in Statistics as the first problem. Basically, we have a sample of size ùëõ from the ùí©(ùúá, ùúé
2

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Define parameters
Œ∏_vals = np.arange(-1, 1, 0.1)
M = 1000  # Number of simulations
n = 10  # Sample size
Œ± = 0.05  # Significance level
œÉ = 1  # Standard deviation

# Initialize list to store coverage probabilities
conf_prob = []

# Loop over each Œ∏ value
for Œ∏ in Œ∏_vals:
    counter = 0
    for _ in range(M):
        # Generate n samples from N(Œ∏, œÉ^2)
        x = np.random.normal(Œ∏, œÉ, n)
        # Calculate the lower and upper bounds of the confidence interval
        lower_CI = np.mean(x) - (st.norm.ppf(1 - Œ±/2) * œÉ / np.sqrt(n))
        upper_CI = np.mean(x) + (st.norm.ppf(1 - Œ±/2) * œÉ / np.sqrt(n))
        # Check if Œ∏ is within the confidence interval
        if lower_CI < Œ∏ < upper_CI:
            counter += 1
    # Calculate coverage probability
    conf_prob.append(counter / M)

# Print the coverage probabilities
#print(conf_prob)

# Plot the results
plt.scatter(Œ∏_vals, conf_prob)
plt.ylim(0.6, 1)
plt.xlabel('Œ∏ values')
plt.ylabel('Coverage Probability')
plt.title('Coverage Probability of Confidence Intervals for Different Œ∏ values')
plt.grid(True)
plt.show()
```

### b)

As we observed from the figure that at all ùúá values, the coverage probability of the confidence interval 
is 1 ‚àí ùõº. However, one may ask that whether the coverage probability changes with the sample size 
or not. So, in the following, we check whether the coverage probability changes with the sample size 
by fixing the parameter ùúá at a fixed value. The reader is encouraged to change the value of ùúá and see 
how the coverage probabilities change as a function of ùëõ. In addition, it is to be noted that as the number of replications ùëÄ increases, the coverage probability will converge to the true coverage 
probability.


```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

# Define parameters
Œ∏ = 0  # Fixed Œ∏ value
œÉ = 1  # Standard deviation
Œ± = 0.05  # Significance level
n_vals = np.arange(1, 500, 1)  # Range of sample sizes
M = 1000  # Number of simulations

# Initialize list to store coverage probabilities
conf_prob = []

# Loop over each sample size
for n in n_vals:
    counter = 0
    for _ in range(M):
        # Generate n samples from N(Œ∏, œÉ^2)
        x = np.random.normal(Œ∏, œÉ, n)
        # Calculate the lower and upper bounds of the confidence interval
        lower_CI = np.mean(x) - (st.norm.ppf(1 - Œ±/2) * œÉ / np.sqrt(n))
        upper_CI = np.mean(x) + (st.norm.ppf(1 - Œ±/2) * œÉ / np.sqrt(n))
        # Check if Œ∏ is within the confidence interval
        if lower_CI < Œ∏ < upper_CI:
            counter += 1
    # Calculate coverage probability
    conf_prob.append(counter / M)

# Plot the results
plt.figure(figsize=(12, 8))
plt.plot(n_vals, conf_prob, color='grey')
plt.axhline(1 - Œ±, linestyle='--', color='red')
plt.xlabel('Sample Size (n)')
plt.ylabel('Coverage Probability')
plt.title('Coverage Probability of Confidence Intervals vs. Sample Size')
plt.grid(True)
plt.show()
```
